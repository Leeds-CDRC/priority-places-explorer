```python
import geopandas as gpd
import pandas as pd
```

```python
df = pd.read_csv('../../../Data/PriorityPlacesData/Lower_layer_Super_Output_Areas_(December_2011)_Population_Weighted_Centroids.csv')
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['X'], df['Y']), crs='EPSG:27700')
```

```python
gdf = gdf.to_crs(4326)
gdf['longitude'] = gdf.geometry.x
gdf['latitude'] = gdf.geometry.y
gdf[['longitude', 'latitude', 'objectid', 'lsoa11cd', 'lsoa11nm']].to_csv('../../../Data/PriorityPlacesData/Lower_layer_Super_Output_Areas_(December_2011)_Population_Weighted_Centroids_WGS.csv', index=False)
```
