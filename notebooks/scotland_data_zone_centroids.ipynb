```python
import geopandas as gpd
```

```python
gdf = gpd.read_file('../../../Downloads/SG_DataZoneCent_2011/SG_DataZone_Cent_2011.shp')
```

```python
gdf = gdf.to_crs('EPSG:4326')
```

```python
gdf['longitude'] = gdf.geometry.x
gdf['latitude'] = gdf.geometry.y
```

```python
gdf.to_csv('../../../Data/PriorityPlacesData/SG_DataZones_Centroids_2011.csv', index=False)
```
